# Removing duplicate in titles

```python
data = pd.read_pickle('Pickle/28k_apparel_data')
```

```python
#Removing the products with very few titles
#sorting the data alphabetically
data_sorted = data[data['title'].apply(lambda x : len(x.split())>4)]
data_sorted.shape[0]
```

```python
#Sorting the whole data['title'] aplhabetically to remove duplicates
data_sorted.sort_values('title',inplace=True,ascending=False)
data_sorted.head()
```

```python
print(sum(data_sorted.duplicated('title')))
# Total number of duplicate titles (these titles will be sorted one under another in the list)
```

Some examples of dupliacte titles that differ only in the last few words.
Titles 1:

```python
indices = []
for k,rows in data_sorted.iterrows():
    indices.append(k)
```

```python
#using alphabetically sorted data(sequential data as shown in example above) removing sentences which differ at the end
#These sentences come after one another in a sequential data
asins = []
i = 0
j = 0
num_data_points = data_sorted.shape[0]
while i < num_data_points:
    previous_i = i
    
    #storing the string in form of list of words
    a = data_sorted['title'].loc[indices[i]].split()
        #searching similar products sequntially
    j = i + 1
    
    while j < num_data_points:
        #storing the string in form of list of words                  
        b = data_sorted['title'].loc[indices[j]].split()
        
        #store max len amongst both the strings
        length = max(len(a),len(b))

        #count is used to store the number of words common in both strings
        count = 0
        
        for h in itertools.zip_longest(a,b):
               
            if (h[0]==h[1]):
                count +=1
        # if the number of words in which both strings differ are > 2 , we are considering it as those two apperals are different
        # if the number of words in which both strings differ are < 2 , we are considering it as those two apperals are same, hence we are ignoring them
        if (length-count) > 2:
            asins.append(data_sorted['asin'].loc[indices[i]])      
        
            #start searching for similar apparels in the next string
            i=j
            break
        else:
            j +=1
    if previous_i == i: #this is when all points are the same
        break        
```

```python
#keeping points in data which are present in our asins list
data = data[data['asin'].isin(asins)]
```

```python
#we removed sequential duplicates using this method
data.shape[0]
```

```python
indices = []
for i,row in data.iterrows():
    indices.append(i)

asins1 = []
while len(indices)!=0:
    i = indices.pop()
    asins1.append(data['asin'].loc[i])
    # consider the first apperal's title
    a = data['title'].loc[i].split()
    # store the list of words of ith string in a list
    for j in indices:
        
        b = data['title'].loc[j].split()
        # store the list of words of jth string in a list
        
        length = max(len(a),len(b))
        
        # count is used to store the number of words that are matched in both strings
        count  = 0

        # itertools.zip_longest(a,b): will map the corresponding words in both strings, it will appened None in case of unequal strings
        # example: a =['a', 'b', 'c', 'd']
        # b = ['a', 'b', 'd']
        # itertools.zip_longest(a,b): will give [('a','a'), ('b','b'), ('c','d'), ('d', None)]
        for h in itertools.zip_longest(a,b): 
            if (h[0]==h[1]):
                count += 1

        # if the number of words in which both strings differ are < 3 , we are considering it as those two apperals are same, hence we are ignoring them
        if (length - count) < 3:
            indices.remove(j)
```

```python
data = data[data['asin'].isin(asins1)]
```

## Text Preprocessing  

```python
data = pd.read_pickle('Pickle/16k_apparel_data')
```

```python
data.head()
```

```python
# we make a list of stopwords from nltk stopwords lib
stop_words = set(stopwords.words('english'))
print(list(stop_words))
```

```python
def preprocessing(text,index,column):
    if type(text) is not int:
        string = ""
        for word in text.split():
            #removing special characters
            words = ("".join(i for i in word if i.isalnum()))
            #making all the words as lower case
            words = words.lower()
            #removal of stop words
            if words not in stop_words:
                string += words + " "
        data[column][index] = string     
```

```python
for index,rows in data.iterrows():
    preprocessing(rows['title'],index,'title')
```

# Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
title_vectorizer = CountVectorizer()
title_feature = title_vectorizer.fit_transform(data['title'])
# CountVectorizer().fit_transform(data['title']) returnsa sparse matrix with dimensions  =  no. points * words_in_corpus

title_feature.shape
```

```python
def bag_of_words_model(doc_id,cut_off):
    #doc_id is the number on the new index formed after CountVectorizer is applied to the data['title']
    #pairwise distance will save the distance between given input apparel and all other apparels
    pairwise_dist = pairwise_distances(title_feature,title_feature[doc_id])
    
    #np.argsort will return indices of the smallest distances
    indice = np.argsort(pairwise_dist.flatten())[0:cut_off]
    
    #get the index of the original dataframe
    data_indice = list(data.index[indice])
    
    for i in data_indice:
        print('ASIN :',data['asin'].loc[i])
        print ('Title:', data['title'].loc[i])

    
```
